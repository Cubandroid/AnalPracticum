```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats
```

```python
def toFixed(numObj, digits=1):
    return f"{numObj:.{digits}f}"
```

```python
pFs = np.array([1.0, 2, 3, 4, 5, 6])
Es = ([-574, -507, -448, -393, -354, -336])
fig = plt.figure(figsize=(15, 10))

ax = fig.add_subplot(111)
ax.set_title(' ', fontsize = 20)
ax.set_xlabel('pF', fontsize=15)
ax.set_ylabel(', ', fontsize = 15)
ax.plot(pFs, Es, 'bo', label = ' F')
slope, intercept, r_value, p_value, std_err = stats.linregress(pFs[0:4],Es[0:4])
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(pFs[4:6],Es[4:6])
ax.plot(pFs, pFs*slope + intercept, label = '     (1 < pF < 4.7), tg = '+ str(toFixed(slope)) + '')
ax.plot(pFs, pFs*slope1 + intercept1, label = '     ')

x = (intercept1-intercept)/(slope-slope1)
controls = np.array([-447, -573])
controlFs = (controls-intercept)/slope
ax.plot(controlFs, controls, 'gv', label = ' ')
ax.plot(x, x*slope + intercept, 'rs', label = ' , pF = ' + toFixed(x))
ax.legend()
```
