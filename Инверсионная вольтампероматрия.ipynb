```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats
import pandas as pd
from scipy.signal import find_peaks
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111)

frame = pd.read_csv('IVA_dataset.csv')
#frame.plot(kind='scatter',x='E, mV',y='I, uA',color='blue')
E = frame['E, mV'].to_numpy()
I = frame['I, uA'].to_numpy()
ax.set_xlabel('E, мВ', fontsize = 15)
ax.set_ylabel('I, мкА', fontsize = 15)
ax.plot(E, I, 'bo')
peaks = find_peaks(I, width = 2)
metals = ['Zn', 'Cd', 'Pb', 'Cu']
j = 0
for i in peaks[0]:
    plt.plot(E[i], I[i], 'ro', label = metals[j])
    ax.annotate(metals[j], xy=(E[i], I[i]), color = 'red', fontsize = 20, xytext=(E[i]-15, I[i]+0.5))
    j+=1

#ax.legend()
```
